```python
from Bio import SeqIO


perGO_file = 'peroxisome_related_genes/uniprot-goa-per.fasta'
glyGO_file = 'peroxisome_related_genes/uniprot-goa-gly.fasta'
per_file = 'peroxisome_related_genes/uniprot-peroxisome.fasta'
gly_file = 'peroxisome_related_genes/uniprot-glyoxysome.fasta'
neg_file = 'uniprot-taxonomy_58024+reviewed_yes.fasta'

output_file = 'raw_neg.fasta'

exclude = set()

glyGO_sequences = SeqIO.parse(open(glyGO_file),'fasta')
for fasta in glyGO_sequences:
    exclude.add(fasta.id)

per_sequences = SeqIO.parse(open(per_file),'fasta')
for fasta in per_sequences:
    exclude.add(fasta.id)
    
gly_sequences = SeqIO.parse(open(gly_file),'fasta')
for fasta in gly_sequences:
    exclude.add(fasta.id)
perGO_sequences = SeqIO.parse(open(perGO_file),'fasta')
for fasta in perGO_sequences:
    exclude.add(fasta.id)

fasta_sequences = SeqIO.parse(open(neg_file),'fasta')
with open(output_file, 'w') as output_handle:
    total_seq = 0
    filtered_seq_no=0
    for fasta in fasta_sequences:
        total_seq += 1
        if fasta.id not in list(exclude) and len(fasta.seq) > 20:
            filtered_seq_no += 1
            SeqIO.write([fasta], output_handle, "fasta")
print("Total peroxisome related protein : {}".format(len(exclude)))
print("Total non-peroxisome protein : {}".format(total_seq))
print("Total added non-peroxisome protein : {}".format(filtered_seq_no))
```
